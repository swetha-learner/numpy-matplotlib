## Distance between 2 points in 3D space

 calculate the dist btwn 2 points in 3D space using arrays. take values of coordinators from user.

```python
import numpy as np
a = np.zeros(3)
b= np.zeros(3)
for num in range(2):
    print("enter the coordinates of point  {}".format(num+1))
    for i in range(3):
        if num==0:
            a[i] = float(input("enter {0} coordinate value:".format(chr(ord('X')+i))))
        else:
            b[i] = float(input("enter {0}coordinate value:".format(chr(ord('X')+i))))
    print("***********************")
print(a)
print(b)
ss=0
for j in range(3):
    ss= ss+(a[j]-b[j])**2
print("the distance betwween these two points is {0:.3f}".format(np.sqrt(ss)))
```

## Inverse of matrix:

calculate inverse of 2x2 matrix and take inputs from the user.

```python
import numpy as np
a = np.zeros((2,2))
for i in range(2):
    for j in range(2):
        a[i][j]=float(input("enter value at {0},{1}:".format(i,j)))
print(a)
print("********************")
b = np.zeros((2,2))
diag=1
ndiag=1
for i in range(2):
    for j in range(2):
        if i==j:
            #print("diagonal")
            b[i][j] = a[1-i][1-j]
            diag = diag*a[i][j]
        else:
            #print("non-diagonal")
            b[i][j]= -a[i][j]
            ndiag= ndiag*a[i][j]
inv = b/ (diag-ndiag)
print("the inverse matrix is:",inv)
```

## 3 phase sine wave plot

```python
import numpy as np
import matplotlib.pyplot as plt
x= np.linspace(0,2*np.rad2deg(np.pi),100)
y1= np.sin(np.deg2rad(x))
y2= np.sin(np.deg2rad(x-120))
y3= np.sin(np.deg2rad(x-240))
plt.figure(dpi=120)
plt.plot(x,y1,color='r',linewidth=3,label="phase1")
plt.plot(x,y2,color="g",linewidth=3,label="phase2")
plt.plot(x,y3,color='b',linewidth=3,label="phase3")
plt.xlabel("phase angle in degrees")
plt.ylabel("units")
plt.title("3 phase plot")
plt.legend(loc=1)
plt.grid(alpha=.5,color="r")
plt.axis([0,360,-1.2,1.2])
plt.show()
```
